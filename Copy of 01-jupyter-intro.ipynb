### Cells
A notebook is divided up into what are called *cells*, which are blocks that contain information like code or text. In Google Colab, there are two types of cells:
- **Text cell**: Has formatted text, images, links, and equations using Markdown syntax. Colab renders the formatting into a readable webpage-like format
- **Code cell**: Has a play button (▶) on the left side that you can click to run the code. By default, code cells contain executable Python code and display any output below the cell. You can also run shell commands by starting the line with ! (e.g. !ls to list files). Magic commands, indicated by the usage of % or %%, allow special operations like changing directories (%cd) or running R code (%R for single line, %%R for multi-line R code through rpy2)

#### Running a cell
"Code" cells can be "run", or executed. If the cell is a "Text" cell, it will automatically be rendered to look like a webpage. A "Code" cell will run the code using a language set by the *runtime* (which we'll get to next).

This next cell is a Code cell. ### Runtimes
The **Runtime** is a virtual machine provided by Google that executes your notebook’s code. You can change the runtime’s settings to suit your needs, such as enabling a GPU or increasing memory. The runtime environment also determines which programming languages are available, by default, Colab uses Python 3.

If you run code that isn’t valid Python, you’ll get an error. To use a different language, you’ll either need to change the runtime type or install the language manually, something we’ll do in later GL4U notebooks.

To change a notebook's runtime type, go to **Runtime > Change Runtime Type** in the top menu. In the dialog that appears, you can select a different programming language and choose hardware acceleration options like GPU or TPU.

Colab currently supports **Python 3**, **R**, and **Julia** as programming language options.
